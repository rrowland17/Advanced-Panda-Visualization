```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
sdc = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
sdc.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
school_count = len(sdc["school_name"].unique())

#Calculate the total number of students
student_count = (sdc["student_name"]).count()

#Calculate the total budget
budget = school_data["budget"].sum()
                 
#Calculate the average math score
avg_math = student_data[("math_score")].mean()

#Calculate the average reading score
avg_read = student_data[("reading_score")].mean()

#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
opr = ((avg_read + avg_math)/2)

#Calculate the percentage of students with a passing math score (70 or greater)
# used apply and lambda, ended up using .loc later on so I could configure data better
pm = student_data.apply(lambda x: True if x["math_score"] >= 70 else False, axis=1)
pm_count = len(pm[pm == True].index)
pm_perc = (pm_count / student_count) * 100


#Calculate the percentage of students with a passing reading score (70 or greater)
pr = student_data.apply(lambda x: True if x["reading_score"] >= 70 else False, axis=1)
pr_count = len(pr[pr == True].index)
pr_perc = (pr_count / student_count) * 100

#Create a dataframe to hold the above results
district_df = pd.DataFrame({"Total Schools": [school_count],
                            "Total Students": [student_count], 
                            "Total Budget": [budget],
                            "Average Math Score": [avg_math], 
                            "Average Reading Score": [avg_read], 
                            "% Passing Math": [pm_perc],
                            "% Passing Reading": [pr_perc],
                            "% Overall Passing Rate": [opr]})
#Optional: give the displayed data cleaner formatting
```

## School Summary

```python
#Create an overview table that summarizes key metrics about each school
#setting school name and type as index
sch_groups = school_data.set_index(["school_name"])["type"]
sch_groups
```

```python
#Total Students
total_student = sdc["school_name"].value_counts()
total_student
```

```python
total_budget = sdc.groupby(["school_name"])["budget"].mean()
total_budget
```

```python
#Per Student Budget
psb = total_budget / total_student
psb
```

```python
#Average Math Score
avg_m = sdc.groupby(["school_name"])["math_score"].mean()
avg_m
```

```python
#Average Reading Score
avg_r = sdc.groupby(["school_name"])["reading_score"].mean()
avg_r
```

```python
#% Passing Math
#changed to .loc functions
ppm = sdc.loc[sdc["math_score"] >= 70]
group_ppm = ppm["school_name"].value_counts()
ppm_perc = (group_ppm / total_student) * 100
ppm_perc
```

```python
#% Passing Reading
ppr = sdc.loc[sdc["reading_score"] >= 70]
group_ppr = ppr["school_name"].value_counts()
ppr_perc = (group_ppr / total_student) * 100
ppr_perc
```

```python
#overall passing percentage
oppr = ((ppm_perc + ppr_perc)/2)
oppr
```

```python
school_summary = pd.DataFrame({"School_Type": sch_groups,
                               "Total Students":total_student,
                               "Total School Budget":total_budget, 
                               "Per Student Budget":psb,
                               "Average Math Score":avg_m,
                               "Average Reading Score":avg_r,
                               "% Passing Math":ppm_perc,
                               "% Passing Reading":ppr_perc,
                               "% Overall Passing Rate":oppr})
school_summary.head()
```
